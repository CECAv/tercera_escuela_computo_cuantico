<a id="1.1"></a>
## 1.1. Computacin cuntica, cmo la uso?

La computacin cuntica es una tecnologa emergente que usa las leyes de la mecnica cuntica (**superposicin, interferencia y entrelazamiento**) para resolver problemas complejos para las computadoras clsicas. Los dispositivos que realizan estos clculos se le llama hardware cuntico.

Al igual que en la computacin clsica, podemos crear **algoritmos cunticos** para estas nuevas computadoras. Una forma popular de implementarlos es mediante circuitos cunticos, que requieren **compuertas cunticas** para su definicin.

<a id="1.2"></a>
## 1.2. En dnde estamos actualmente?

Las computadoras cunticas actuales son pequeas (pocos qubits) e inestables, sin superar a las tradicionales.

Hay dos factores limitantes en los problemas que pueden resolver:

1. **Cantidad de qubits**: Un **qubit** es un *bit cuntico*. Si no tenemos suficientes qubits, no podemos almacenar ni operar problemas grandes.
2. **Tasa de error**: Las computadoras cunticas actuales son ruidosas y frecuentemente se equivocan, introduciendo "*ruido*" (informacin intil).

Por ejemplo, para factorizar un nmero grande (617 dgitos) en menos de un da, se estiman necesarios ~20 millones de qubits ruidosos. Hasta julio de 2024, IBM ha creado una computadora de 1,121 qubits y planea sistemas ms robustos. Otros algoritmos podran ofrecer ventaja cuntica antes, como se menciona en el artculo de IBM: "[Evidence for the utility of quantum computing before fault tolerance](https://www.nature.com/articles/s41586-023-06096-3)".

Empresas estn invirtiendo en algoritmos y computacin cuntica para aprovechar esta ventaja.

Cualquiera puede crear y ejecutar algoritmos cunticos para aprender a usar esta tecnologa emergente.

# 5. Qubit

Al igual que en las computadoras digitales estndar (computadoras clsicas) la unidad fundamental son los **bits**, en las computadoras cunticas tenemos a los **qubits**, una extensin del bit a la mecnica cuntica.

Una definicin ms precisa de **qubit** sera:

<div class="alert alert-block alert-success">
Un qubit es un sistema mecnico cuntico de dos estados (o dos niveles) que pertenece a un espacio de Hilbert bidimensional ($\mathbb{C}^2$).

Un [espacio de Hilbert](https://en.wikipedia.org/wiki/Hilbert_space) es un [espacio vectorial](https://en.wikipedia.org/wiki/Vector_space) que define un producto interior. El espacio vectorial ocupado para definir a los qubits utiliza el campo de los nmeros complejos $\mathbb{C}$ y como deseamos que sea bidimensional, es que se expresa como $\mathbb{C}^2$, es decir, necesitamos 2 nmeros complejos para describir a un **qubit**.

### Recordatorio: Nmeros complejos

Un nmero complejo $\mathbb{C}$ es un elemento de un sistema numrico que ampla los nmeros reales con un elemento especfico denotado con $i$, llamado **unidad imaginaria** y satisface la ecuacin 

$$i^2 = 1$$

todo nmero complejo se puede expresar en la forma

$$z=a + bi$$

donde $a$ y $b$ son nmeros reales, teniendo as una *parte real* ($a$) y otra *parte imaginaria* ($b$). 

As $a,b\in\mathbb{R}$, mientras que $z\in\mathbb{C}$.

<a id="5.1"></a>
## 5.1. Representacin de los qubits: Matemticamente

Para representar a los qubits de manera matemtica, se utiliza la **[notacin de Dirac](https://en.wikipedia.org/wiki/Bra%E2%80%93ket_notation)**, tambin conoida como notacin *bra-ket*.

El estado $\ket{0}$ se pronuncia "**ket 0**", y el estado $\ket{1}$ es "**ket 1**". Esto dos estados (o qubits) son ortogonales entre s y forman una base $\{\ket{0}, \ket{1}\}$, llamada **base computacional**, debido a que abarcan el [espacio vectorial](https://en.wikipedia.org/wiki/Vector_space) lineal bidimensional (espacio de Hilbert, $\mathbb{C}^2$) del qubit.

La notacin de Dirac es una manera corta de escribir **vectores**. Hemos dicho que un qubit es un estado de dos niveles, y para representar qu tanto tenemos de cada nivel, necesitamos un vector de dos dimensiones. La manera ms general de escribir un estado cuntico (un quibit) es entonces:

$$\ket{\psi}=\mymatrix{r}{ \alpha \\ \beta }$$

donde $\alpha,\beta\in\mathbb{C}$ y el ket $\ket{\psi}$ es la forma corta de escribir el estado de dos niveles (el vector).

En el caso de los elementos de la base computacional $\{\ket{0}, \ket{1}\}$, representan a los *vectores de estado* siguientes:

$$\ket{0}=\mymatrix{r}{ 1 \\ 0 } ~~~~~~~~~~~~~~~~~~~~ \ket{1}=\mymatrix{r}{ 0 \\ 1 }$$

Entonces, un qubit, es una superposicin de los estados base, es decir, un qubit puede ser expresado como la [combinacin lineal](https://en.wikipedia.org/wiki/Linear_combination) de $\ket{0}$ y $\ket{1}$:

$$\ket{\psi}=\alpha\ket{0}+\beta\ket{1}$$

donde $\alpha$ y $\beta$ son [amplitudes de probabilidad](https://en.wikipedia.org/wiki/Probability_amplitude). Cuando medimos este qubit, en la base computacional, la probabiliadad de obtener el estado $\ket{0}$ es de $|\alpha|^2$ y la probabilidad de obtener el estado $\ket{1}$ es de $|\beta|^2$. Debido a que los cuadrados de los valores absolutos de las amplitudes equivalen a probabilidades, se deduce que $\alpha$ y $\beta$ deben estar restringidos de acuerdo con el [segundo axioma de la teora de la probabilidad](https://en.wikipedia.org/wiki/Probability_axioms#Second_axiom) mediante la ecuacin:

$$|\alpha|^2 + |\beta|^2 = 1$$

**Nota**: Dada esta restricin, notemos que la longitud de los vectores siempre ser igual a $1$.

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>:

$$\ket{\psi}=\frac{\sqrt{3}}{2}\ket{0}+\frac{1}{2}\ket{1}=\mymatrix{c}{ \dfrac{\sqrt{3}}{2} \\\\ \dfrac{1}{2} }$$

La probabilidad de observar el estado $\ket{0}$ es de $\left|\dfrac{\sqrt{3}}{2}\right|^2=\dfrac{3}{4}$, que es $75\%$.
    
La probabilidad de observar el estado $\ket{1}$ es de $\left|\dfrac{1}{2}\right|^2=\dfrac{1}{4}$, que es $25\%$.
    
La suma de las probabilidades para cada caso: $\dfrac{3}{4}+\dfrac{1}{4}=1$, que es $100\%$.

### Recordatorio: Base de un espacio vectorial

Un conjunto $B$ de vectores en un espacio vectorial $V$ se llama **base** si cada elemento de $V$ puede escribirse de manera nica como una combinacin lineal finita de elementos de $B$.

De manera equivalente, un conjunto $B$ es una base si sus elementos son linealmente independientes y cada elemento de $V$ es una combinacin lineal de elementos de $B$. En otras palabras, una base es un *conjunto generador* linealmente independiente.

### Otras bases para los qubits

La base ms comn es la previamente mencionada, la base computacional, definida por los vectores: $\{\ket{0}, \ket{1}\}$.

Pero podemos usar cualesquiera otros dos vectores para definir una base para los qubits, siempre y cuando, sean linealmente independientes, pertenezcan al espacio vectorial, en este caso el espacio de $\mathbb{C}^2$ (dos nmeros complejos).

Sin embargo, hay bases que sirven para entablar una comunicacin entre quienes hacemos comptacin cuntica, y las bases con elementos ortogonales son preferidas.

Otra base comn es $\{\ket{+}, \ket{-}\}$, donde la definicin de estos vectores es,

$$\ket{+}=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} } ~~~~~~~~~~~~~~~~~~~~ \ket{-}=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} }$$

Estos estados tambin pueden ser escritos en trminos de la base computacional como,

$$\ket{+}=\dfrac{1}{\sqrt2}\Big(\ket{0}+\ket{1}\Big)$$

$$\ket{-}=\dfrac{1}{\sqrt2}\Big(\ket{0}-\ket{1}\Big)$$

Notemos que los coeficientes de ambas bases, son nmeros reales, deben pertenecer a los complejos, pero en particular ahora tenemos que $z=a+0i$, por lo que queda solo la parte real (siguen siendo nmeros dentro del campo $\mathbb{C}$).

Entonces es fcil realizar una grfica de estos cuatro vectores:

## 5.2. Representacin de los qubits: Esfera de Bloch

Qu pasa si las amplitudes de los qubits ya no son solo nmeros reales?

Entonces sern nmeros complejos, como se mencion anteriormente, es decir $b\neq 0$, teniendo as tanto la parte real como la imaginaria: $z=a+bi$.

Debido a que un qubit se escribe,

$$\ket{\psi}=\mymatrix{r}{ \alpha \\ \beta }$$

con $\alpha,\beta\in\mathbb{C}$, es decir, necesita dos nmeros complejos, y a su vez cada nmero complejo necesita de 2 nmeros reales para poder escribirlo, entonces, necesitamos en total $4$ nmeros reales para representar a un qubit.

$$\alpha=a+bi$$

$$\beta=c+di$$

donde $a,b,c,d \in \mathbb{R}$.

Podra parecer a primera vista que se tienen $4$ grados de libertad para un qubit, sin embargo, un grado de libertad deja de ser *libre* debido a la restriccin de normalizacin: $|\alpha|^2 + |\beta|^2 = 1$. Tenemos ahora 3 grados de libertad.

Entonces, utilizando una *transformacin* de coordenadas conveniente, se puede eliminar ese grado de libertad, es decir, lograr la conversin de $\mathbb{R}^4\rightarrow\mathbb{R}^3$. Una eleccin posible son las [coordenadas de Hopf](https://en.wikipedia.org/wiki/3-sphere#Hopf_coordinates), que manda el espacio de $4$ dimensiones a la superficie de una esfera unitaria en $3$ dimensiones.

Despus de considerar que para un qubit, una *fase global* del estado no tiene consecuencias fsicamente observables, se obtiene la **esfera de Bloch** (ms detalles [aqu](https://en.wikipedia.org/wiki/Qubit#Bloch_sphere_representation)).

<div class="alert alert-block alert-success">
La esfera de Bloch es una representacin geomtrica de un sistema mecnico cuntico de dos niveles (un quibit).

<p style="font-size:10px;font-style:italic;text-align:right"><a href="https://en.wikipedia.org/wiki/Bloch_sphere" target="_blank">Imagen obtenida de Wikipedia</a></p>

Entonces, el estado cuntico, en trminos de la esfera de Bloch, se escribe como,

$$\begin{aligned}
\ket{\psi}&=\cos\left(\frac{\theta}{2}\right) \, \ket{0}+e^{i\varphi}\sin\left(\frac{\theta}{2}\right) \, \ket{1}\\\\\\
&=\cos\left(\frac{\theta}{2}\right)\mymatrix{r}{ 1 \\ 0 } + e^{i\varphi}\sin\left(\frac{\theta}{2}\right)\mymatrix{r}{ 0 \\ 1 }\\\\
&=\mymatrix{r}{ \cos\left(\frac{\theta}{2}\right) \\ e^{i\varphi}\sin\left(\frac{\theta}{2}\right) }
\end{aligned}$$

donde 

$$0\leq\theta\leq\pi$$

$$0\leq\varphi < 2\pi$$

Entonces, la transformacin es,

$$\alpha=\cos\left(\frac{\theta}{2}\right)$$

$$\beta=e^{i\varphi}\sin\left(\frac{\theta}{2}\right)$$

Notemos que bajo esta transformacin, $\alpha\in\mathbb{R}$ y $\beta\in\mathbb{C}$.

$$$$

**Nota**: Los parmetros $\theta$ y $\varphi$, respectivamente pueden ser reinterpretados en coordenadas esfricas como la [colatitud](https://en.wikipedia.org/wiki/Colatitude) con respecto al eje $z$ y la [longitud](https://en.wikipedia.org/wiki/Longitude) con respecto al eje $x$. Es decir, $(x,y,z)=(\sin\theta\cos\varphi, \sin\theta\sin\varphi, \cos\theta)$ con $r=1.$

$$$$

En la esfera de Bloch, los [puntos antipodales](https://en.wikipedia.org/wiki/Antipodal_point) corresponden a un par de vectores de estado mutuamente ortogonales. El *polo norte* y el *polo sur* de la esfera de Bloch generalmente se eligen para que correspondan a los vectores base estndar $\ket{0}$ y $\ket{1}$ (cannicos), respectivamente, que a su vez podra corresponder, por ejemplo, a los estados de 'espn arriba' y 'espn abajo' de un electrn, a veces escritos como: $\ket{\uparrow}$ y $\ket{\downarrow}$. Sin embargo, esta **eleccin es arbitraria**.

Entonces, los estados $\ket{0}$ y $\ket{1}$ en los polos norte y sur, son **espacialmente antiparalelos** (lo que observamos en la esfera de Bloch), pero **en el espacio de Hilbert son ortogonales**.

En efecto, es un poco confuso tener dos nociones diferentes de "ortogonal", una para el espacio fsico y otra para el espacio de Hilbert, pero esto viene del hecho de tener dos espacios diferentes en los que debemos pensar.

### Otras bases en la esfera Bloch

<p style="font-size:10px;font-style:italic;text-align:right"><a href="https://en.wikipedia.org/wiki/File:Six_states_Bloch_sphere.jpg" target="_blank">Imagen obtenida de Wikipedia</a></p>

La base formada por $\{\ket{0}, \ket{1}\}$ son puntos antipodales sobre el eje $z$. La base formada por $\{\ket{+}, \ket{-}\}$ son puntos antipodales sobre el eje $x$. Entonces podemos esperar que exista una base que est situada en el eje $y$, y as es, dicha base est formada por los estados $\{\ket{+i}, \ket{-i}\}$, que en trminos de la base computacional se escriben,

$$\ket{+i}=\dfrac{1}{\sqrt2}\Big(\ket{0}+i\ket{1}\Big)=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} }$$

$$\ket{-i}=\dfrac{1}{\sqrt2}\Big(\ket{0}-i\ket{1}\Big)=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ -\frac{i}{\sqrt{2}} }$$

$$$$

**Nota**: Estas tres bases son las ms comunes, pero no son las nicas, mientras un par de estados, o vectores, sirvan para generar el espacio, pueden formar una base es decir, que no sean linealmente dependientes en el espacio de Hilbert.

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>: Veamos que los elementos de la base $\{\ket{+i}, \ket{-i}\}$ efectivamente son ortogonales.
    
$$$$
    
Dos vectores son ortogonales cuando su producto interno es igual a cero, en notacin de Dirac, se escribe como un **braket**, es decir: $\braket{\phi}{\psi}$, en donde el bra $\bra{\phi}$ indica el complejo conjugado transpuesto del vector $\ket{\phi}$, entonces tenemos que

$$$$
    
$$\braket{+i}{-i}=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} }^* \; \mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{-i}{\sqrt{2}} }=\mymatrix{r}{ \frac{1}{\sqrt{2}} & \frac{-i}{\sqrt{2}} }\cdot\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{-i}{\sqrt{2}} }=\left(\frac{1}{\sqrt{2}}\right)\left(\frac{1}{\sqrt{2}}\right)+\left(\frac{-i}{\sqrt{2}}\right)\left(\frac{-i}{\sqrt{2}}\right)=\frac{1}{2}-\frac{1}{2}=0$$

En el siguiente ejemplo se define un circuito cuntico con 3 qubits, los cuales no tienen que ser creados necesariamente usando la clase `QuantumRegister`, sino que basta con indicar la cantidad de qubits que deseamos tenga el circuito, tambin se definen 2 bits clsicos (lneas dobles), y se aplican 3 compuertas. Como se mencion, el tiempo ocurre de izquierda a derecha, entonces la compuerta $H$ es la que se aplica primero, y despus la compuerta $CNOT_{0,1}$ y por ltimo la compuerta $CNOT_{0,2}$, al final ocurren las mediciones.

# 7. Compuerta Cuntica

Una vez que tenemos qubits en un circuito cuntico, lo siguiente es manipularlos, para lograr un objetivo, implementar algoritmos cunticos. Esta manipulacin de hace con las **compuertas cunticas**.

<div class="alert alert-block alert-success">
Las compuertas cunticas son operadores unitarios y se describen como matrices unitarias en relacin con alguna base. Usualmente se usa la base computacional. 

Las compuertas cunticas son los componentes bsicos de los circuitos cunticos, como lo son las compuertas lgicas clsicas para los circuitos digitales convencionales. A diferencia de muchas compuertas lgicas clsicas, las compuertas lgicas cunticas son **reversibles**.

### Recordatorio: Matriz Unitaria

En lgebra lineal, una matriz cuadrada compleja $U$ es unitaria si su transpuesta conjugada $U^*$ es tambin su inversa, es decir, 

$$U^*U=UU^*=UU^{-1}=I$$

donde $I$ es la matriz identidad.

En mecnica cuntica, la transpuesta conjugada se conoce como el [adjunto hermitiano](https://en.wikipedia.org/wiki/Hermitian_adjoint) de una matriz y se denota con una daga ($\dagger$), por lo que la ecuacin anterior se escribe,

$$U^{\dagger}U=UU^{\dagger}=I$$

Las matrices unitarias tienen una importancia significativa en la mecnica cuntica porque conservan normas (longitudes) y, por lo tanto, amplitudes de probabilidad.

<a id="7.1"></a>
## 7.1. Representacin de las compuertas: matemticamente

Una compuerta cuntica que acta sobre $n$ qubits, esta representada por una matriz unitaria de $2^n\times 2^n$. Los vectores que representan estados cunticos de dos niveles (qubits) sobre los que actan las compuertas son vectores unitarios en $2^n$ dimensiones.

Las compuertas cunticas ms comunes operan en espacios vectoriales de uno o dos qubits, al igual que las compuertas lgicas clsicas comunes operan en uno o dos bits.

La accin de una compuerta en un estado cuntico especfico se encuentra multiplicando el vector $\ket{\psi_1}$ que representa el estado, por la matriz $U$ que representa la compuerta. El resultado es un nuevo estado cuntico $\ket{\psi_2}$:

$$U\ket{\psi_1}=\ket{\psi_2}$$

Cuando tenemos un qubit, $n=1$, el vector que lo representa tiene dos elementos y una compuerta que opera sobre un qubit es una matriz de $2\times 2$, obtener el estado resultante de aplicar la compuerta ser como sigue,

$$U\ket{\psi_1}=\mymatrix{rr}{ u_{11} & u_{12} \\ u_{21} & u_{22} }\mymatrix{r}{ \alpha \\ \beta }=\mymatrix{r}{ \alpha \, u_{11} + \beta \, u_{12} \\ \alpha \, u_{21} + \beta \, u_{22} }=\mymatrix{r}{ \alpha' \\ \beta' }=\ket{\psi_2}$$

### 7.2.1. Compuertas de Pauli

Las **compuertas de Pauli** son las tres [matrices de Pauli](https://en.wikipedia.org/wiki/Pauli_matrices) ($\sigma_x, \sigma_y,\sigma_z$) y actan sobre un solo quibit. 

Las compuertas $X$, $Y$, $Z$, respectivamente, equivalen a una rotacin de $\pi$ radianes (180) al rededor del eje $x$, $y$, $z$ en la *esfera de Bloch*.

### Compuerta $X$

Esta compuerta es equivalente a la compuerta de negacin ($NOT$) en las computadoras clsicas.

Tambin es llamada *bit-flip* ya que mapea los estados base como sigue:

$$X\ket{0} = \ket{1}$$

$$X\ket{1} = \ket{0}$$

Su representacin matricial es,

$$X=NOT=\mymatrix{rr}{ 0 & 1 \\ 1 & 0 }$$

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>: Efecto de la compuerta $X$ sobre el estado $\ket{1}$
    
$$X\ket{1}=\mymatrix{rr}{ 0 & 1 \\ 1 & 0 }\mymatrix{r}{ 0 \\ 1 }=\mymatrix{r}{ 1 \\ 0 }=\ket{0}$$

En Qiskit aplicamos la compuerta $X$ con el mtodo `x()`:

````python
circ1 = QuantumCircuit(1) # circuito con 1 qubit

circ1.x(0)                # aplicar compuerta X al (nico) qubit 0

circ1.draw('mpl')         # mostrar el circuito
````

### Obtener el vector de estado con Qiskit

Con la clase `Statevector` de Qiskit, se puede obtener el estado de un circuito en el punto de se desee:

````python
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex


circ1 = QuantumCircuit(1)    # circuito con 1 qubit

psi_0 = Statevector(circ1)   # estado justo despus de crear el circuito

circ1.x(0)                   # aplicar compuerta X al (nico) qubit 0

psi_1 = Statevector(circ1)   # estado despus aplicar la compuerta X

display(circ1.draw('mpl'))   # mostrar el circuito

print(" Estado inicial:")
print(" en forma de vector:")
display(array_to_latex(psi_0.data))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus de X:")
print(" en forma de vector:")
display(array_to_latex(psi_1.data))
print(" en forma de ket:")
display(psi_1.draw('latex'))
````

Esta compuerta mapea los estados base como sigue:

$$Y\ket{1}=-i\ket{0}$$

Su representacin matricial es,

$$Y=\mymatrix{rr}{ 0 & -i \\ i & 0 }$$

Con notacin de Dirac:
    
$$Y\ket{+}=Y \, \left(\dfrac{1}{\sqrt2}\Big(\ket{0}+\ket{1}\Big)\right)=\dfrac{1}{\sqrt2}\Big(Y\ket{0}+Y\ket{1}\Big)=\dfrac{1}{\sqrt2}\Big(i\ket{1}-i\ket{0}\Big)=\dfrac{-i}{\sqrt2}\Big(\ket{0}-\ket{1}\Big)=-i\ket{-}$$

Con notacin matricial:
    
$$Y\ket{+}=\mymatrix{rr}{ 0 & -i \\ i & 0 }\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} }=\mymatrix{r}{ \frac{-i}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} }=-i\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{-1}{\sqrt{2}} }=-i\ket{-}$$

````python
circ2 = QuantumCircuit(1)    # circuito con 1 qubit

psi_0 = Statevector(circ2)   # estado justo despus de crear el circuito

circ2.y(0)                   # aplicar compuerta Y al (nico) qubit 0

psi_1 = Statevector(circ2)   # estado despus aplicar la compuerta Y

display(circ2.draw('mpl'))   # mostrar el circuito

print(" Estado inicial:")
print(" en forma de vector:")
display(array_to_latex(psi_0))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus de Y:")
print(" en forma de vector:")
display(array_to_latex(psi_1))
print(" en forma de ket:")
display(psi_1.draw('latex'))
````

### 7.2.2. Rotaciones con otros ngulos

Las compuertas de Pauli son rotaciones de $\pi$ radianes, pero podemos hacer que el estado cuntico rote con otros ngulos, para ello, tenemos las compuertas de rotaciones, alrededor de los ejes, definidas como sigue:

$$R_x(\phi)=\mymatrix{rr}{ \cos\left(\frac{\phi}{2}\right) & -i\sin\left(\frac{\phi}{2}\right) \\ -i\sin\left(\frac{\phi}{2}\right) & \cos\left(\frac{\phi}{2}\right) }=\cos\left(\frac{\phi}{2}\right) \, I-i\sin\left(\frac{\phi}{2}\right) \, X$$

$$R_y(\phi)=\mymatrix{rr}{ \cos\left(\frac{\phi}{2}\right) & -\sin\left(\frac{\phi}{2}\right) \\ \sin\left(\frac{\phi}{2}\right) & \cos\left(\frac{\phi}{2}\right) }=\cos\left(\frac{\phi}{2}\right) \, I-i\sin\left(\frac{\phi}{2}\right) \, Y$$

$$R_z(\phi)=\mymatrix{cc}{ e^{-i\frac{\phi}{2}} & 0 \\ 0 & e^{i\frac{\phi}{2}} }=\cos\left(\frac{\phi}{2}\right) \, I-i\sin\left(\frac{\phi}{2}\right) \, Z$$

En Qiskit existen los mtodos `rx()`, `ry()` y `rz()` para lograr las rotaciones con diferentes ngulos.

En el siguiente cdigo rotaremos el estado cuntico primero $\pi/2$ radianes al rededor dej eje $y$, lo que lo dejar sobre el eje $+x$, despus lo rotaremos $\pi/4$ radianes alrededor del eje $z$ y finalmente $\pi$ radianes alrededor del eje $x$. En forma matemtica, expresamos estos pasos como:

$$\ket{\psi_3}=R_x\left(\pi\right) \, R_z\left(\frac{\pi}{4}\right) \, R_y\left(\frac{\pi}{2}\right) \; \ket{0}$$

Notemos que el orden al multiplicar las matrices ocurre de derecha a izquierda, es decir, se multiplica primero la matriz que est ms cerca del estado $\ket{0}$, despus la que sigue a la izquierda y por ltimo la que aparece al principio en la lectura de la expresin.

Los pasos son entonces:

$\ket{\psi_0}=\ket{0} ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ =\ket{0}$

$\ket{\psi_1}= R_y\left(\frac{\pi}{2}\right) \; \ket{\psi_0} ~~~~~~~~~~~~~~ = R_y\left(\frac{\pi}{2}\right) \; \ket{0}$

$\ket{\psi_2}= R_z\left(\frac{\pi}{4}\right) \; \ket{\psi_1} ~~~~~~~~~~~~~~ = R_z\left(\frac{\pi}{4}\right) \, R_y\left(\frac{\pi}{2}\right) \; \ket{0}$

$\ket{\psi_3}=R_x\left(\pi\right) \; \ket{\psi_2} ~~~~~~~~~~~~~~~~ = R_x\left(\pi\right) \, R_z\left(\frac{\pi}{4}\right) \, R_y\left(\frac{\pi}{2}\right) \; \ket{0}$

````python
import numpy as np


circ4 = QuantumCircuit(1)    # circuito con 1 qubit

psi_0 = Statevector(circ4)   # estado justo despus de crear el circuito

circ4.ry(np.pi/2, 0)         # rotar pi/2 alrededor del eje y

psi_1 = Statevector(circ4)   # estado despus de la primera rotacin

circ4.rz(np.pi/4, 0)         # rotar pi/4 alrededor del eje z

psi_2 = Statevector(circ4)   # estado despus de la segunda rotacin

circ4.rx(np.pi, 0)           # rotar pi alrededor del eje x

psi_3 = Statevector(circ4)   # estado despus de la tercera rotacin

display(circ4.draw('mpl'))   # mostrar el circuito

print(" Estado inicial:")
print(" en forma de vector:")
display(array_to_latex(psi_0))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus rotacin de pi/2 alrededor del eje y:")
print(" en forma de vector:")
display(array_to_latex(psi_1))
print(" en forma de ket:")
display(psi_1.draw('latex'))
print()
print(" Estado despus rotacin de pi/4 alrededor del eje z:")
print(" en forma de vector:")
display(array_to_latex(psi_2))
print(" en forma de ket:")
display(psi_2.draw('latex'))
print()
print(" Estado despus rotacin de pi alrededor del eje x:")
print(" en forma de vector:")
display(array_to_latex(psi_3))
print(" en forma de ket:")
display(psi_3.draw('latex'))

print()
print(" Estado inicial:")
display(plot_bloch_multivector(psi_0))

print(" Estado despus rotacin de pi/2 alrededor del eje y:")
display(plot_bloch_multivector(psi_1))

print(" Estado despus rotacin de pi/4 alrededor del eje z:")
display(plot_bloch_multivector(psi_2))

print(" Estado despus rotacin de pi alrededor del eje x:")
display(plot_bloch_multivector(psi_3))
````

### 7.2.3. Compuerta Hadamard

La compuerta Hadamard crea un estado de superposicin de la base computacional $\{\ket{0}, \ket{1}\}$, entonces mapea los estados como sigue,

$$H\ket{0}=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} }=\ket{+}$$

$$H\ket{1}=\mymatrix{r}{ \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} }=\ket{-}$$

$$$$

La compuerta Hadamard tambin es una rotacin, pero alrededor del eje $\frac{\left(\hat{x}+\hat{z}\right)}{\sqrt{2}}$, es decir, el eje en el plano $xz$ con 45 del estado $\ket{0}$

<p style="font-size:10px;font-style:italic;text-align:right"><a href="https://physics.stackexchange.com/questions/313959/visual-interpretation-on-the-bloch-sphere-when-hadamard-gate-is-applied-twice" target="_blank">Imagen obtenida de Physics StackExchange</a></p>

Su representacin matricial es,

$$H=\frac{1}{\sqrt{2}} \mymatrix{rr}{1 & 1 \\ 1 & -1} = \mymatrix{rr}{\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}}$$

<div class="alert alert-block alert-info">
<strong>Ejemplo</strong>: Efecto de la compuerta $H$ sobre el estado $\ket{-}$

Con notacin de Dirac:
    
$$\begin{aligned}
H\ket{-}&=H \, \left(\dfrac{1}{\sqrt2}\Big(\ket{0}-\ket{1}\Big)\right)=\dfrac{1}{\sqrt2}\Big(H\ket{0}-H\ket{1}\Big)\\
&=\dfrac{1}{\sqrt2}\left[\left(\dfrac{1}{\sqrt2}\Big(\ket{0}+\ket{1}\Big)\right)-\left(\dfrac{1}{\sqrt2}\Big(\ket{0}-\ket{1}\Big)\right)\right]\\
&=\dfrac{1}{2}\Big(\ket{0}+\ket{1}-\ket{0}+\ket{1}\Big)=\dfrac{1}{2}\Big(2\ket{1}\Big)=\ket{1}
\end{aligned}$$
    
Con notacin matricial:
    
$$H\ket{-}=\mymatrix{rr}{\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}}\mymatrix{rr}{ \frac{1}{\sqrt{2}} \\ -\frac{1}{\sqrt{2}} }=\frac{1}{2}\mymatrix{r}{ 1-1 \\ 1+1 }=\frac{1}{2}\mymatrix{r}{ 0 \\ 2 }=\mymatrix{r}{ 0 \\ 1 }=\ket{1}$$

En Qiskit aplicamos la compuerta $H$ con el mtodo `h()`:

````python
circ5 = QuantumCircuit(1)    # circuito con 1 qubit

psi_0 = Statevector(circ5)   # estado justo despus de crear el circuito

circ5.h(0)                   # apliquemos H al estado |0>, quedamos en |+>

psi_1 = Statevector(circ5)   # estado justo despus de aplicar H

circ5.z(0)                   # aplicar compuerta Z al (nico) qubit 0, quedamos en el estado |->

psi_2 = Statevector(circ5)   # estado despus aplicar la compuerta Z

circ5.h(0)                   # aplicar compuerta H de nuevo, quedamos en el estado |1>

psi_3 = Statevector(circ5)   # estado despus aplicar la compuerta Z

display(circ5.draw('mpl'))   # mostrar el circuito

print(" Estado inicial:")
print(" en forma de vector:")
display(array_to_latex(psi_0))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus de primera H:")
print(" en forma de vector:")
display(array_to_latex(psi_1))
print(" en forma de ket:")
display(psi_1.draw('latex'))
print()
print(" Estado despus de Z:")
print(" en forma de vector:")
display(array_to_latex(psi_2))
print(" en forma de ket:")
display(psi_2.draw('latex'))
print()
print(" Estado despus de segunda H:")
print(" en forma de vector:")
display(array_to_latex(psi_3))
print(" en forma de ket:")
display(psi_3.draw('latex'))

print()
print(" Estado inicial:")
display(plot_bloch_multivector(psi_0))

print(" Estado despus de primera H:")
display(plot_bloch_multivector(psi_1))

print(" Estado despus de Z:")
display(plot_bloch_multivector(psi_2))

print(" Estado despus de segunda H:")
display(plot_bloch_multivector(psi_3))
````

### 7.2.4. Compuertas de cambio de fase

Las compuertas de cambio de fase forman una familia de compuertas de un solo qubit que mapean los estados base como sigue,

$$P(\phi) \, \ket{0} = \ket{0}$$

$$P(\phi) \, \ket{1} = e^{i\phi}\ket{1}$$

La probabilidad de medir $\ket{0}$ o $\ket{1}$ no cambia despus de aplicar esta compuerta, sin embargo, modifica la fase del estado cuntico. Esto es equivalente a trazar un crculo horizontal (una lnea de latitud), o una rotacin a lo largo del eje $z$ en la esfera de Bloch por $\phi$ radianes. La compuerta de cambio de fase est representada por la matriz:

$$P(\phi)=\mymatrix{rr}{1 & 0 \\ 0 & e^{i\phi}}$$

donde $\phi$ es el cambio de fase con periodo $2\pi$.

Notemos que la compuerta $P(\phi)$ no es [hermitiana](https://en.wikipedia.org/wiki/Hermitian_matrix) (excepto para todos los $\phi=n\pi, \; n\in\mathbb{Z}$).

Existen compuertas con su propio nombre para algunos casos particulres del valor de $\phi$.

$$$$

**Compuertas $S$ y $S^{\dagger}$**

Es una rotacin alrededor del eje $z$ por $\pi/2$ radianes, entonces

$$S=\sqrt{Z}=P\left(\frac{\pi}{2}\right)$$

La compuerta $S^{\dagger}$ es el conjugado de la compuerta $S$, entonces,

$$S^{\dagger}=P\left(-\frac{\pi}{2}\right)$$

Sus matrices son,

$$S=\mymatrix{rr}{1 & 0 \\ 0 & i} ~~~~~~~~~~~~~~~~~~~~~~~~~~~ S^{\dagger}=\mymatrix{rr}{1 & 0 \\ 0 & -i}$$

$$$$

**Compuertas $T$ y $T^{\dagger}$**

Es una rotacin alrededor del eje $z$ por $\pi/4$ radianes, entonces

$$T=\sqrt{S}=\sqrt[4]{Z}=P\left(\frac{\pi}{4}\right)$$

$$T^{\dagger}=P\left(-\frac{\pi}{4}\right)$$

Sus matrices son,

$$T=\mymatrix{rr}{1 & 0 \\ 0 & e^{i\frac{\pi}{4}}} ~~~~~~~~~~~~~~~~~~~~~~~~~~~ T^{\dagger}=\mymatrix{rr}{1 & 0 \\ 0 & e^{-i\frac{\pi}{4}}}$$

En Qiskit tenemos a nuestra disposicin mtodos para aplicar las compuertas de cambio de fase: `p()`, `s()`, `sdg()`, `t()`, `tdg()`.

Veamos el efecto de estas compuertas, pero debido a que representan rotaciones alrededor del eje $z$, entonces partiremos del estado $\ket{+}$, para que sean notorias la rotaciones.

````python
circ6 = QuantumCircuit(1)    # circuito con 1 qubit

circ6.h(0)                   # nos movemos al estado |+>

psi_0 = Statevector(circ6)   # estado justo despus de crear el circuito y posicionarse en |+>

circ6.s(0)                   # aplicar compuerta S, rotando pi/2

psi_1 = Statevector(circ6)   # estado justo despus de aplicar P

circ6.tdg(0)                 # aplicar compuerta T rotanfo -pi/4

psi_2 = Statevector(circ6)   # estado despus aplicar la compuerta S

circ6.p(2*np.pi/3, 0)        # aplicar compuerta P, rotando 2pi/3

psi_3 = Statevector(circ6)   # estado despus aplicar la compuerta T

display(circ5.draw('mpl'))   # mostrar el circuito

print(" Estado inicial |+>:")
print(" en forma de vector:")
display(array_to_latex(psi_0))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus de S:")
print(" en forma de vector:")
display(array_to_latex(psi_1))
print(" en forma de ket:")
display(psi_1.draw('latex'))
print()
print(" Estado despus de T:")
print(" en forma de vector:")
display(array_to_latex(psi_2))
print(" en forma de ket:")
display(psi_2.draw('latex'))
print()
print(" Estado despus de P(2*pi/3):")
print(" en forma de vector:")
display(array_to_latex(psi_3))
print(" en forma de ket:")
display(psi_3.draw('latex'))

print()
print(" Estado inicial |+>:")
display(plot_bloch_multivector(psi_0))

print(" Estado despus de S:")
display(plot_bloch_multivector(psi_1))

print(" Estado despus de T:")
display(plot_bloch_multivector(psi_2))

print(" Estado despus de P(2*pi/3):")
display(plot_bloch_multivector(psi_3))
````

### 7.2.5. Compuerta ms general de un qubit

La compuerta que generaliza todas las compuertas de un solo qubit es la conocida como $U$ y est definida como:

$$U(\theta, \phi,\lambda)=\mymatrix{rr}{\cos\left(\frac{\theta}{2}\right) & -e^{i\lambda}\sin\left(\frac{\theta}{2}\right) \\ e^{i\phi}\sin\left(\frac{\theta}{2}\right) & e^{i\left(\phi+\lambda\right)}\cos\left(\frac{\theta}{2}\right)}$$

(en otras referencias tambin la puedes encontrar como compuerta $U3$)

ejemplos de equivalencias:

$$U\left(\theta, -\frac{\pi}{2}, \frac{\pi}{2}\right)=RX(\theta)$$

$$U\left(\theta, 0, 0\right)=RY(\theta)$$

En Qiskit podemos usar el mtodo `u()` para aplicar esta compuerta:

````python
circ_u = QuantumCircuit(1)               # circuito con 1 qubit

psi_0 = Statevector(circ_u)              # estado justo despus de crear el circuito e iniciar el qubit en |1>

circ_u.u(np.pi/3, np.pi/2, np.pi/5, 0)   # aplicar compuerta U3 al (nico) qubit en la posicin 0

psi_1 = Statevector(circ_u)              # estado despus aplicar la compuerta U3

display(circ_u.draw('mpl'))              # mostrar el circuito

print(" Estado inicial:")
print(" en forma de vector:")
display(array_to_latex(psi_0))
print(" en forma de ket:")
display(psi_0.draw('latex'))
print()
print(" Estado despus de U3(pi/3, pi/8, pi/5):")
print(" en forma de vector:")
display(array_to_latex(psi_1))
print(" en forma de ket:")
display(psi_1.draw('latex'))

print()
print(" Estado inicial:")
display(plot_bloch_multivector(psi_0))

print(" Estado despus de U3(pi/3, pi/8, pi/5):")
display(plot_bloch_multivector(psi_1))
````
